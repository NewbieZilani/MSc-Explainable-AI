```python
df = pd.read_csv("weather_dataset_2025.csv")

df.head()
```

```python
df.info()
```

```python
# seperate numeric columns
numeric_df = df.select_dtypes(include="number")

numeric_df.head()
```

```python
# create numpy array
weather = numeric_df.to_numpy()

print(weather)
```

```python
# extract the column
temperature = weather[:, 0]

humidity = weather[:, 1]

wind = weather[:, 2]

precipitation = weather[:, 3]
```

```python
print(temperature[:5])
```

```python
print(weather.shape)
rows, columns = weather.shape

print("Rows :", rows)
print("Columns :", columns)
print(weather.size)
```

```python
# Indexing & Slicing
print(weather[:5])
print(weather[0])
print(weather[3])

# specific element finding
print(weather[0,0])
print(weather[3,2])

# To get full column
temperature=weather[:,0]
print(temperature)
print(temperature.size)

wind = weather[:, 2]
print(wind)
print(wind.size)
```

```python
# Row
print(weather[1])
# first 5th row
print(weather[:5])
print(weather[10:15])
```

```python
# Column Slicing
# every row but only temperature and Humidity
print(weather[:, 0:2])

# only wind and Precipitation
print(weather[:, 2:4])

# first 10 rows and first 2 column
print(weather[0:10, 0:2])

# first 5 rows and last 2 column
print(weather[:5, 2:])
```
