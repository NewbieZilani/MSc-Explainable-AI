```python
import numpy as np
print(np.__version__)
```

```python
marks = np.array([85, 72, 90, 65, 88, 92, 75, 80, 95, 70])
print(marks)
```

```python
print(np.mean(marks))
print(np.max(marks))
print(np.min(marks))
print(np.sum(marks))
```

```python
marks2d = np.array([
    [80,70,90,60],
    [75,85,88,91],
    [95,90,92,94],
    [60,65,70,68],
    [88,84,82,80]
])

print(marks2d)
```

```python
# row=student, column=subject
# all marks for 1 student
print(marks2d[0])
print(marks2d[1])
```

```python
# all student for 1 subject
print(marks2d[:,0])
print(marks2d[:,3])
```

```python
# 1 subject for 1 student
print(marks2d[2,2])
print(marks2d[3,1])
```

```python
# average for 1 student
print(np.mean(marks2d[0]))
print(np.mean(marks2d[2]))
```

```python
# average for all students
print(np.mean(marks2d, axis=1))
# average for single subject
print(np.mean(marks2d, axis=0))
```
