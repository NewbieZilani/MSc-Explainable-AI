```python
covid_df["date"]=pd.to_datetime(covid_df["date"])
covid_df.info()
```

```python
totalcases=covid_df["new_cases"].sum()
print(totalcases)
```

```python
totalDeath=covid_df["new_deaths"].sum()
print(totalDeath)
```

```python
# Maximum death in a single day
covid_df.loc[covid_df["new_deaths"].idxmax()]
```

```python
# highest death in a single day without all row
covid_df["new_deaths"].max()
```

```python
# Maximum cases in a single day
covid_df.loc[covid_df["new_cases"].idxmax()]
```

```python
covid_df["new_cases"].max()
```

```python
# top 10 day when new cases was maximum
covid_df.sort_values("new_cases", ascending=False).head(10)
```

```python
# top 10 day when new deaths was maximum
covid_df.sort_values("new_deaths", ascending=False).head(10)
```

```python
# total cases in a month
covid_df["month"]=covid_df["date"].dt.month
covid_df.groupby("month")["new_cases"].sum()
```
